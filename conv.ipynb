```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

def exp_decay(x, a, b, c):
    return a * np.exp(-b * x) + c

x = np.linspace(-5, 5, 1000)
y_gaussian = gaussian(x, 0, 1)
y_exp_decay = exp_decay(x, 1, 2, 0.5)
y_conv = np.convolve(y_gaussian, y_exp_decay, mode='same') * (x[1]-x[0])

plt.plot(x, y_conv, label='Convolution')
plt.legend()

plt.show()
```
